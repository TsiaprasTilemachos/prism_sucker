# Histogram of Oriented Gradients

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
image = cv.imread(r"C:\Users\ttsiapras\Documents\REASEARCH_MATERIAL\IMAGE_PROCESS\images\lenna.bmp",0)
W,H   = image.shape
_ = plt.imshow(image,cmap="gray"),plt.axis("off"),plt.title("Original")
```

```python
image = cv.copyMakeBorder(image, 1, 1, 1, 1, cv.BORDER_REFLECT)
_ = plt.imshow(image,cmap="gray"),plt.axis("off"),plt.title("Padded [x1]")
```

```python
image = image.astype(int)
Dx   = np.zeros((W,H)).astype(int)
Dy   = np.zeros((W,H)).astype(int)
Dmag = np.zeros((W,H)).astype(int)
for i in range(1,H+1):
    for j in range(1,W+1):
        Dx[i-1,j-1] = image[i,j+1]-image[i,j-1]
        Dy[i-1,j-1] = image[i-1,j]-image[i+1,j]
        Dmag[i-1,j-1]=math.sqrt(Dx[i-1,j-1]**2 + Dy[i-1,j-1]**2)
fig = plt.figure(figsize=(10,5))
fig.add_subplot(1,3,1)
plt.imshow(Dx,cmap="gray"),plt.axis("off"),plt.title("DX")
fig.add_subplot(1,3,2)
plt.imshow(Dy,cmap="gray"),plt.axis("off"),plt.title("DY")
fig.add_subplot(1,3,3)
plt.imshow(Dmag,cmap="gray"),plt.axis("off"),plt.title("Dmag")
```
